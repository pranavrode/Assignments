## Task 6 : Business wants to check the Claim Amount for deviation for each transaction, they would like you to calculate as follow CLAIM_DEVIATION = AVG_CLAIM_AMOUNT_FOR_LAST_30DAYS (same insurance type) / CURRENT_CLAIM_AMOUNT. If the value < 0.5 THEN CLAIM_DEVIATION = 1 ELSE 0. If there is less than 30 days of transaction history THEN -1 Note: LAST_30DAYS does not include current day

```python
view.TXN_DATE_TIME.max()
```

```python
# Assuming current day to be 2021-06-30
curr_date = pd.to_datetime('2021-06-30')
```

```python
view['TXN_DATE_TIME'] = pd.to_datetime(view['TXN_DATE_TIME'])
```

```python
## df with last 30 days data
last_30 = view.loc[(view['TXN_DATE_TIME'] > '2021-05-30') & (view['TXN_DATE_TIME'] < '2021-06-30')]  
```

```python
# Getting average claim of last 30 days insurance type wise and storing it in dictionary
average_claim = last_30.groupby('INSURANCE_TYPE').mean()['CLAIM_AMOUNT'].to_dict()
```

```python
# defining a function for calculating claim deviation
def Claim_dev(row):
    if (curr_date - row['TXN_DATE_TIME'])<pd.Timedelta(days=30):
        return -1
    claim_deviation = average_claim[row['INSURANCE_TYPE']] / row['CLAIM_AMOUNT']
    if claim_deviation < 0.5:
        return 1
    else:
        return 0
```

```python
# Applying the function to the dataframe
view['CLAIM_DEVIATION'] = view.apply(Claim_dev,axis=1)
```

```python
view.CLAIM_DEVIATION.value_counts()
```

There are 9216 claims with deviation less than 0.5

784 claims with transaction history of less than 30 days
